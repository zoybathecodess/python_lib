# Merge
### combining two dataframes based on a common column (key)

```python
lotr.merge(lotr2) #default is inner
```

```python
lotr.merge(lotr2, how = 'inner', on = ['FirstName', 'FellowshipID']) #inner merges on the basis of common columns 
```

```python
lotr.merge(lotr2, how = 'outer') #returns everything from both the dataframes 
```

```python
lotr.merge(lotr2, how = 'left') #takes everything from left dataframe (lotr in this case), if anything overlaps/matches in the right df, merges the two and returns
```

```python
lotr.merge(lotr2, how = 'right')
```

```python
lotr.merge(lotr2, how = 'cross') #compares each row of the left df with the right df 
```

# Join 

```python
lotr.join(lotr2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
df = lotr.set_index('FellowshipID').join(lotr2.set_index('FellowshipID'), lsuffix='_left', rsuffix= '_right', how = 'outer')
df
```

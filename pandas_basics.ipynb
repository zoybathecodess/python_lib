## adding/removing columns using lambda function

```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x<165 else('Average' if x<185 else 'Tall'))
```

```python
bios.tail()
```

```python
def weight_category(row):
    if row['height_cm']<175 and row['weight_kg']<75:
        return 'LightWeight'
    elif row['height_cm']<195 or row['weight_kg']<80:
        return 'MediumWeight'
    else:
        return 'HeavyWeight'

bios['Category'] = bios.apply(weight_category,axis=1)
bios
```
